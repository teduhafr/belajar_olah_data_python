```python
sales_df[sales_df.total_price.isna()]
```

```python
sales_df["total_price"] = sales_df["price_per_unit"] * sales_df["quantity"]
```

```python
# SETELAH INI AKAN MELAKUKAN EKPLORASI DATA
```

```python
customers_df.describe(include="all")
```

```python
#GROUP PELANGGAN BERDASARKAN JENIS KELAMIN
customers_df.groupby(by="gender").agg({
    "customer_id": "nunique",
    "age": ["max", "min", "mean", "std"]
})
```

```python
#MELIHAT SEBARAN KOTA
customers_df.groupby(by="state").customer_id.nunique().sort_values(ascending=False)
```

```python
customers_df.groupby(by="city").customer_id.nunique().sort_values(ascending=False)
```

```python
#EXPLORE DATA ORDERS
delivery_time = orders_df["delivery_date"] - orders_df["order_date"] #ngitung delivery time, akakn membuat kolom baru delivery time
delivery_time = delivery_time.apply(lambda x: x.total_seconds()) #dihitung per detiknya
orders_df["delivery_time"] = round(delivery_time/86400)  #dijadikan hari (86400 detik / hari)
```

```python
orders_df.head()
```

```python
orders_df.describe(include="all")
```

```python
#buat tabel baru di customer, status -> aktif yang pernah order, kalau yang belum pernah non aktif
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

```python
#terlihat banyak pelanggan yang belum pernah melakukan order
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
#cek jumlah order berdasarkan kota
```

```python
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
#order berdasarkan state
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

```python
#order berdasarkan gender
orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

```python
#order berdasarkan kelompok usia
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
```

```python
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
#Eksplorasi Data product_df dan sales_df
product_df.describe(include="all")

```

```python
sales_df.describe(include="all")
```

```python
product_df.sort_values(by="price", ascending=False)
```

```python
#pivot produk berdasarkan tipe dan nama
```

```python
product_df.groupby(by="product_type").agg({
    "product_id": "nunique",
    "quantity": "sum",
    "price":  ["min", "max"]
})
 
product_df.groupby(by="product_name").agg({
    "product_id": "nunique",
    "quantity": "sum",
    "price": ["min", "max"]
})
```

```python
sales_product_df = pd.merge(
    left=sales_df,
    right=product_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
sales_product_df.head()
```

```python
sales_product_df.groupby(by="product_type").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
#cek penjualan berdasarkan nama produk
```

```python
sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

```python
#Eksplorasi Data all_df
all_df = pd.merge(
    left=sales_product_df,
    right=orders_customers_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_df.head()
```

```python
#cek preferensi pembelian
all_df.groupby(by=["state", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
#cek produk berdasarkan gender dan kelompok usia
all_df.groupby(by=["gender", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
all_df.groupby(by=["age_group", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
monthly_orders_df.head()
```
